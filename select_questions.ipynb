```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# set dpi to 300 rcparams
plt.rcParams['figure.dpi'] = 300

import json
RANDOM_SEED = 729
```

```python
with open('Data/certainty/all_permutations/t_[1, 2, 3]_three_probs,two_probs_Control.json') as f:
    data = json.load(f)
```

```python
# make data into a dataframe where each row is the key and the values are the values
df = pd.DataFrame(list(data.items()), columns=['key', 'values'])
# expand the dictionary in the values column into a dataframe
df = pd.DataFrame(df['values'].apply(pd.Series))
df = pd.concat([df, df['subtemplates'].apply(pd.Series)], axis=1)
```

```python
# select ten random rows from the dataframe, set seed
n_samples = 10
df_sample = df.groupby('permutation_index').sample(n_samples // 2, random_state=RANDOM_SEED).reset_index()
```
